# Ploting the Data Points with Regression Line

```python
import matplotlib.pyplot as plt
from scipy import stats
from scipy import stats
```

```python
x =[5,7,8,7,2,17,2,9,4,11,12,9,6]
y =[99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
slope, intercept,r,p,std_err =stats.linregress(x, y)
```

```python
print("slope:", slope, " intercept:", intercept, " r:", r, " p:", p, " stderr:", std_err)
```

```python
def myfunc(x):
    return slope * x + intercept
```

```python
mymodel =list(map(myfunc, x))
```

```python
plt.scatter(x, y)
plt.plot(x, mymodel)
```

# Relationship in LR :

```python
slope,intercept,r,p,std_err = stats.linregress(x, y)
print("The relationship coefficient is:",r)
```

The result -0.76 shows that there is a relationship, not perfect, but it indicates that 
we could use linear regression in future predictions.
